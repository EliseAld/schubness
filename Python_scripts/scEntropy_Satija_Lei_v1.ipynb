```python
import scEntropy.scEntropy as scEntropy 
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_file_path = '/Users/elise/Desktop/Imagine_bckp/scRNAseq/Data_sc/hubness_data/10X/GSE100866_CBMC_8K_13AB_10X-RNA_umi.csv'
satija_data = pd.read_csv(data_file_path, index_col=0, header=0)
```

```python
satija_scentropy = scEntropy.scEntropy(satija_data, ref_vec=None, option='predefined')
```

```python
plt.plot(satija_scentropy)
plt.show;
```
