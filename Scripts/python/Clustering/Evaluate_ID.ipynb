```python
rootdir = '../data/DuoClustering2018/sce_filteredM3Drop10/'
resdir = '../results/DuoClustering2018/sce_filteredM3Drop10/'

fnames = os.listdir(rootdir)
fnames = [f for f in fnames if 'sce' in f and '.rds' in f]

gid = {}
lid = {}
for n_comps in [25,50,100]:#,500]:
    gid[n_comps] = {}
    lid[n_comps] = {}
    print(n_comps)
    for m in ['fisher','twonn','pca','danco']:
        gid[n_comps][m] = {}
        lid[n_comps][m] = {}
    
    caldata = skdim.id.DANCo(fractal=False).fit(np.random.random((100,n_comps))).calibration_data_
    for fname in fnames:
        print('\t\t'+fname)
        s=time.time()
        adata = readRDS(rootdir+fname)
        adata.X = adata.layers['logcounts']
        sc.pp.pca(adata, n_comps = min(adata.X.shape[1]-1,min(len(adata.X)-1,n_comps)))
        X = adata.obsm['X_pca']
        dists, inds = skdim.get_nn(X,100,n_jobs=4)
        
        gid[n_comps]['fisher'][fname] = skdim.id.FisherS().fit(X).dimension_
        gid[n_comps]['twonn'][fname] = skdim.id.TwoNN().fit(X).dimension_
        gid[n_comps]['pca'][fname] = skdim.id.lPCA(alphaFO=.1).fit(X).dimension_
        gid[n_comps]['danco'][fname] = skdim.id.DANCo(fractal=False).fit(X).dimension_
        
        lid[n_comps]['fisher'][fname] = skdim.asPointwise(X,skdim.id.FisherS(),precomputed_knn=inds,n_jobs=4)
        lid[n_comps]['twonn'][fname] = skdim.asPointwise(X,skdim.id.TwoNN(),precomputed_knn=inds,n_jobs=4)
        lid[n_comps]['pca'][fname] =  skdim.asPointwise(X,skdim.id.lPCA(alphaFO=.1),precomputed_knn=inds,n_jobs=4)
        lid[n_comps]['danco'][fname] = skdim.asPointwise(X,skdim.id.DANCo(fractal=False,calibration_data=caldata),precomputed_knn=inds,n_jobs=4)
        print(round((time.time()-s)/60,2))
        %xdel adata
        gc.collect()
        
with open(resdir+f'all_id{n_comps}.pkl','wb') as f:
    pickle.dump([gid,lid],f)
```

```python
import scipy
def get_csr(f,shape):
    #shape = (n_dims, n_samples)
    mat = benchmark_common_functions.IncrementalCSRMatrix(shape, np.float32)

    for row_ind,line in enumerate(f,start=-1):
        print(row_ind,end='\r')
        if row_ind==-1:
            continue

        data = np.array(line[1:],dtype=float)
        idx = data.nonzero()[0]
        for col_ind in idx:
            mat.append(row_ind, col_ind, data[col_ind])
            
    return mat


base_folder = '../data/scRNAseq_Benchmark_datasets/Intra-dataset/'
base_folder2 = '../data/scRNAseq_Benchmark_datasets/Inter-dataset/'
resdir = '../results/scRNAseq_Benchmark_datasets/'

get_res_path = lambda fname: f'{resdir}{fname}_norm{do_norm}_scale{norm_scale}_pca{do_pca}_ncomps{n_comps}_{metric}_weighted{weighted}_knn{n_neighbors}_seed{seed}.pkl'

all_data_paths = dict(
    #AMB
    AMB3={'data':base_folder+'AMB/Filtered_mouse_allen_brain_data.csv',
         'labels':base_folder+'AMB/Labels.csv',
         'labels_key':'Class'},
    
    AMB16={'data':base_folder+'AMB/Filtered_mouse_allen_brain_data.csv',
         'labels':base_folder+'AMB/Labels.csv',
         'labels_key':'Subclass'},
    
    AMB92={'data':base_folder+'AMB/Filtered_mouse_allen_brain_data.csv',
         'labels':base_folder+'AMB/Labels.csv',
         'labels_key':'cluster'},
    
    #CellBench
    cellbench1={'data':base_folder+'CellBench/10x_5cl/10x_5cl_data.csv',
                'labels':base_folder+'CellBench/10x_5cl/Labels.csv',
                'labels_key':'x'},
    
    cellbench2={'data':base_folder+'CellBench/CelSeq2_5cl/CelSeq2_5cl_data.csv',
                'labels':base_folder+'CellBench/CelSeq2_5cl/Labels.csv',
                'labels_key':'x'},

    #Pancreatic
    BaronHuman={'data':base_folder+'Pancreatic_data/Baron Human/Filtered_Baron_HumanPancreas_data.csv',
            'labels':base_folder+'Pancreatic_data/Baron Human/Labels.csv',
            'labels_key':'x'},
    BaronMouse={'data':base_folder+'Pancreatic_data/Baron Mouse/Filtered_MousePancreas_data.csv',
                'labels':base_folder+'Pancreatic_data/Baron Mouse/Labels.csv',
                'labels_key':'x'},
    Muraro={'data':base_folder+'Pancreatic_data/Muraro/Filtered_Muraro_HumanPancreas_data.csv',
                'labels':base_folder+'Pancreatic_data/Muraro/Labels.csv',
           'labels_key':'x'},
    Segerstolpe={'data':base_folder+'Pancreatic_data/Segerstolpe/Filtered_Segerstolpe_HumanPancreas_data.csv',
                'labels':base_folder+'Pancreatic_data/Segerstolpe/Labels.csv',
                'labels_key':'x'},
    Xin={'data':base_folder+'Pancreatic_data/Xin/Filtered_Xin_HumanPancreas_data.csv',
                'labels':base_folder+'Pancreatic_data/Xin/Labels.csv',
            'labels_key':'x'},
    #TM
    TM={'data':base_folder+'TM/Filtered_TM_data.csv',
        'labels':base_folder+'TM/Labels.csv',
        'labels_key':'x'},
    #Zheng
    Zhengsorted={'data':base_folder+'Zheng sorted/Filtered_DownSampled_SortedPBMC_data.csv',
    'labels':base_folder+'Zheng sorted/Labels.csv',
    'labels_key':'x'},
    
    Zheng68k={'data':base_folder+'Zheng 68K/Filtered_68K_PBMC_data.csv',
    'labels':base_folder+'Zheng 68K/Labels.csv',
    'labels_key':'x'},
    
    #PbmcBench
    PbmcBench_10Xv2={'data':base_folder2+'PbmcBench/10Xv2/10Xv2_pbmc1.csv',
            'labels':base_folder2+'PbmcBench/10Xv2/10Xv2_pbmc1Labels.csv'},
    PbmcBench_10Xv3={'data':base_folder2+'PbmcBench/10Xv3/10Xv3_pbmc1.csv',
            'labels':base_folder2+'10Xv3/10Xv3_pbmc1Labels.csv'},
    PbmcBench_CELseq={'data':base_folder2+'PbmcBench/CEL-Seq/CL_pbmc1.csv',
            'labels':base_folder2+'PbmcBench/CEL-Seq/CL_pbmc1Labels.csv'},
    PbmcBench_Dropseq={'data':base_folder2+'PbmcBench/Drop-Seq/DR_pbmc1.csv',
            'labels':base_folder2+'PbmcBench/Drop-Seq/DR_pbmc1Labels.csv'},
    PbmcBench_inDrop={'data':base_folder2+'PbmcBench/inDrop/iD_pbmc1.csv',
            'labels':base_folder2+'PbmcBench/inDrop/iD_pbmc1Labels.csv'},
    PbmcBench_SeqWell={'data':base_folder2+'PbmcBench/Seq-Well/SW_pbmc1.csv',
            'labels':base_folder2+'PbmcBench/Seq-Well/SW_pbmc1Labels.csv'},
    
    PbmcBench_SmartSeq2={'data':base_folder2+'PbmcBench/Smart-Seq2/SM2_pbmc1.csv',
            'labels':base_folder2+'PbmcBench/Smart-Seq2/SM2_pbmc1Labels.csv'},
    
    #SCINA data : moignard, Paul (ignored for now)
    #VISP, ALM, MTG
)


fnames=[i for i in all_data_paths.keys() if 'Pbmc' not in i]
#fixed params
do_log = True #already done with do_norm
do_pca = True
n_neighbors = 10
seed = 0

#vary params
n_comps=(25,50,100,500)
do_norm = ('duo','seurat')
norm_scale = (True,False)
params_list = list(product(n_comps,do_norm,norm_scale))

gid = {}
lid = {}
for n_comps,do_norm,norm_scale in params_list:
    print(n_comps,do_norm,norm_scale)
    if f'all_id_pca_abdelaal{n_comps}_{do_norm}_{norm_scale}.pkl' in os.listdir(resdir):
        continue
    gid[(n_comps,do_norm,norm_scale)] = {}
    lid[(n_comps,do_norm,norm_scale)] = {}

    for m in ['fisher','twonn','pca','danco']:
        gid[(n_comps,do_norm,norm_scale)][m] = {}
        lid[(n_comps,do_norm,norm_scale)][m] = {}
    
    caldata = skdim.id.DANCo(D=n_comps,fractal=False).fit(np.random.random((100,n_comps))).calibration_data_
    for fname in fnames:
        print('\t\t'+fname)
        s=time.time()
        if fname == 'TM':
            shape=(54865, 19791)
            adata = anndata.AnnData(X=get_csr(csv.reader(open(all_data_paths[fname]['data'],'r')),shape).tocsr())

        elif fname =='Zheng68k':
            shape=(65943,20387)
            adata = anndata.AnnData(X=get_csr(csv.reader(open(all_data_paths[fname]['data'],'r')),shape).tocsr())

        elif 'AMB' in fname:
            if fname=='AMB3':
                shape= (12832,42625)
                adata = anndata.AnnData(X=get_csr(csv.reader(open(all_data_paths[fname]['data'],'r')),shape).tocsr())
            else:
                gid[(n_comps,do_norm,norm_scale)]['fisher'][fname] =gid[(n_comps,do_norm,norm_scale)]['fisher']['AMB3'] 
                gid[(n_comps,do_norm,norm_scale)]['twonn'][fname] = gid[(n_comps,do_norm,norm_scale)]['twonn']['AMB3']
                gid[(n_comps,do_norm,norm_scale)]['pca'][fname] = gid[(n_comps,do_norm,norm_scale)]['pca']['AMB3']
                gid[(n_comps,do_norm,norm_scale)]['danco'][fname] = gid[(n_comps,do_norm,norm_scale)]['danco']['AMB3']
                lid[(n_comps,do_norm,norm_scale)]['fisher'][fname] =lid[(n_comps,do_norm,norm_scale)]['fisher']['AMB3']
                lid[(n_comps,do_norm,norm_scale)]['twonn'][fname] = lid[(n_comps,do_norm,norm_scale)]['twonn']['AMB3']
                lid[(n_comps,do_norm,norm_scale)]['pca'][fname] =  lid[(n_comps,do_norm,norm_scale)]['pca']['AMB3'] 
                lid[(n_comps,do_norm,norm_scale)]['danco'][fname] = lid[(n_comps,do_norm,norm_scale)]['danco']['AMB3']
                continue
            
        else:
            adata = sc.read_csv(all_data_paths[fname]['data'])
            adata.X = scipy.sparse.csr_matrix(adata.X)
        
        if do_norm == 'seurat':
            benchmark_common_functions.recipe_seurat(adata,do_log,norm_scale)
            print(f'\t\tseurat norm retained {adata.X.shape[1]} genes')
        elif do_norm == 'zheng17':
            benchmark_common_functions.recipe_zheng17(adata,do_log,norm_scale,n_top_genes=5000)
            print(f'\t\tzheng norm retained {adata.X.shape[1]} genes')
        elif do_norm == 'duo':
            benchmark_common_functions.recipe_duo(adata,do_log,renorm=norm_scale)
            print(f'\t\tduo norm retained {adata.X.shape[1]} genes')
        else:
            raise ValueError("do_norm not in 'duo', seurat', 'zheng17'")
        
        sc.pp.pca(adata, n_comps = min(adata.X.shape[1]-1,min(adata.X.shape[0]-1,n_comps)))
        if len(adata) > 2500:
            sc.pp.subsample(adata,n_obs=2500)
        X = adata.obsm['X_pca']
        dists, inds = skdim.get_nn(X,100)
        gid[(n_comps,do_norm,norm_scale)]['fisher'][fname] = skdim.id.FisherS().fit(X).dimension_
        gid[(n_comps,do_norm,norm_scale)]['twonn'][fname] = skdim.id.TwoNN().fit(X).dimension_
        gid[(n_comps,do_norm,norm_scale)]['pca'][fname] = skdim.id.lPCA(alphaFO=.1).fit(X).dimension_
        gid[(n_comps,do_norm,norm_scale)]['danco'][fname] = skdim.id.DANCo(D=n_comps, fractal=False).fit(X).dimension_
        lid[(n_comps,do_norm,norm_scale)]['fisher'][fname] = skdim.asPointwise(X,skdim.id.FisherS(),precomputed_knn=inds)
        lid[(n_comps,do_norm,norm_scale)]['twonn'][fname] = skdim.asPointwise(X,skdim.id.TwoNN(),precomputed_knn=inds)
        lid[(n_comps,do_norm,norm_scale)]['pca'][fname] =  skdim.asPointwise(X,skdim.id.lPCA(alphaFO=.1),precomputed_knn=inds)
        lid[(n_comps,do_norm,norm_scale)]['danco'][fname] = skdim.asPointwise(X,skdim.id.DANCo(D=n_comps,fractal=False,calibration_data=caldata),precomputed_knn=inds)
        print(round((time.time()-s)/60,2))
        %xdel adata
        gc.collect()
        
    with open(resdir+f'all_id_pca_abdelaal{n_comps}_{do_norm}_{norm_scale}.pkl','wb') as f:
        pickle.dump([gid,lid],f)
```
